```python
import numpy as np
from matplotlib.pyplot import *
```

```python
%matplotlib inline
```

```python
def compute_angle_radii(a, b, c):
    c0 = (a*c)/(b - a)
    alpha = np.arctan(a/c0)
#     alpha = np.arctan((b - a)/c)
    d_full = (c + c0) / np.cos(alpha)
    phi = 2 * np.pi * c / d_full
    d0 = c0 / np.cos(alpha)
    chord = d * np.sin(phi / 2) * 2
    return phi, d0, d, chord
```

```python
phi, d0, d, chord = compute_angle_radii(0.7, 0.9, 0.4)
```

```python
phis = np.arange(-2 * np.pi, 2 * np.pi + 0.01, 0.01)
print(phi, phi*180/np.pi, d0, d, chord)
x_ = np.arange(0, d, 0.01)
plot(d*np.cos(phis), d*np.sin(phis))
plot(d0*np.cos(phis), d0*np.sin(phis))
plot(x_*np.cos(phi), x_*np.sin(phi))
plot(x_*np.cos(0), x_*np.sin(0))
axis('equal')
```
